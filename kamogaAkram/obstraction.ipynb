```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    @abstractmethod
    def display_info(self):
        pass

class Car(Vehicle):
    def __init__(self, make, model, year, number_doors):
        super().__init__(make, model, year)
        self.number_doors = number_doors

    def display_info(self):
        return f'Car Make: {self.make}, Model: {self.model}, Year: {self.year}, Doors: {self.number_doors}'

cars = [Car("toyota", "mk4", 2020, 2)]
for car in cars:
    print(car.display_info())
```
